__Creating a data frame of GSVA of all states by merging the .csv files__<br>
1. Assigning the path of the folder containing the GSVA files to a variable 'path'
2. Reading all the files in it using 'glob.glob' into the variable 'all_files'
3. Creating a new variable 'Data_I_B'
4. Using 'for loop' for each file in the folder
5. Reading each file
6. Isolating the data from the columns 'Item' and '2014-15' from the whole data in data frame for each file as we only need the data for the year 2014-15
7. Transposing the data frame to add the the data from each file for the year 2014-15 as a row one after the other as the columns are the same for every file
8. Setting the index as the names of the columns
9. Removing the extra row of column names
10. Adding a new column 'Name' to the data frame which contains the names of each state obtained by deriving the name from the file name using the split function until we get the appropriate name
11. Appending the each row created to the variable 'Data_I_B'
12. Creating a 'master_df' by concatenating the data in the variable 'Data_I_B' by ignoring the index as that creates the column headings for each row and also using 'sort = False' to avoid sorting the columns alphabetically

```python
path = r'C:\Users\dell\Documents\UpGrad\GDP assignment\Data I B'
all_files = glob.glob(path+'/*.csv')
Data_I_B = []
for file in all_files:
    df = pd.read_csv(file, encoding = 'ISO8859')
    df = df[['Item', '2014-15']]
    df = df.T
    df.columns = df.iloc[0]
    df = df[1:]
    df['Name'] = file.split('\\')[-1].split('.')[0].split(' ')[1]
    Data_I_B.append(df)
Data_I_B
```

```python
master_df = pd.concat(Data_I_B, axis = 0, ignore_index=True, sort = False)
master_df
```

```python
# Now the master_df has the GSVA of all states and union territories(Chandigarh, Delhi, Pudducherry)
# So to remove the union territories..

a = master_df.set_index('Name')
a.drop(["Chandigarh"], inplace = True)
a.drop(["Delhi"], inplace = True)
a.drop(["Pudducherry"], inplace = True)
a
```

```python
# Our data frame has got some columns repeated at the end (after the Percapita GSDP in Rs column) due to the * in those
# So we do the analysis after removing those extra columns as they only have values for some states with * in that field

GSVA_states = a.iloc[:,0:-4]
GSVA_states
```

__Plot the percentage contribution of the primary, secondary and tertiary sectors as a percentage of the total GDP for all the states__

```python
# Adding new columns 'Primary_Percent', 'Secondary_Percent', 'Tertiary_Percent' to the original data frame 'GSVA_states'
# Primary_Percent is obtained by dividing the values in 'Primary' with those of 'Gross State Domestic Product'
# Secondary_Percent is obtained by dividing the values in 'Secondary' with those of 'Gross State Domestic Product'
# Tertiary_Percent is obtained by dividing the values in 'Tertiary' with those of 'Gross State Domestic Product'
# Finally multiplying with 100 to get the percentage

warnings.filterwarnings('ignore')
GSVA_states['Primary_Percent'] = 100*(GSVA_states['Primary']/GSVA_states['Gross State Domestic Product'])
GSVA_states['Secondary_Percent'] = 100*(GSVA_states['Secondary']/GSVA_states['Gross State Domestic Product'])
GSVA_states['Tertiary_Percent'] = 100*(GSVA_states['Tertiary']/GSVA_states['Gross State Domestic Product'])
```

__Why is (Primary + Secondary + Tertiary) not equal to total GDP?__

```python
new_df = pd.DataFrame({'Primary': GSVA_states.loc[:,'Primary'],
                    'Secondary': GSVA_states.loc[:,'Secondary'],
                    'Tertiary': GSVA_states.loc[:,'Tertiary'],
                    'Taxes':GSVA_states.loc[:,'Taxes on Products'],
                    'Subsidies':GSVA_states.loc[:,'Subsidies on products'],
                    'GSDP':GSVA_states.loc[:,'Gross State Domestic Product'],
                    'States':list(GSVA_states.index)})
new_df
```

```python
# Lets choose a state('Goa') to perform calculations
# Summing up  the values of Primary, Secondary and Tertiary sectors
# Assigning it to a variable 'Sector_sum'

Sector_sum = new_df.loc['Goa','Primary'] + new_df.loc['Goa','Secondary'] + new_df.loc['Goa','Tertiary']
Sector_sum
```

```python
# The Taxes on the products also add up to the Total revenue generated by the government
# Adding the 'Taxes' to the 'Sector_sum' to obtainn the total 'Revenue'

Revenue = Sector_sum + new_df.loc['Goa','Taxes']
Revenue
```

Revenue is higher than the GSDP and not equal to it because, the government gives away subsidies from the Revenue generated.<br> So we need to subtract the subsidies from the Revenue to get the GSDP. 

```python
# Subtracting the 'Subsidies' from the Revenue generated to get the GSDP

GSDP = Revenue - new_df.loc['Goa','Subsidies']
GSDP
```

```python
new_df.loc['Goa','GSDP']
```

This is why the sum of Primary, Secondary and Tertiary sectors is not equal to the GSDP.

__Find correlation of percentile of the state (% of states with lower per capita GDP) and %contribution of Primary sector to total GDP.__

```python
# Finding percentile of the state and assigning it to a variable 'Percentile_rank'
# Finding percentile of the %contribution of Primary sector to total GDP and assigning it to a variable 'Primary_rank'
# Adding those columns to the original data frame
# Creating a new data frame 'corr_df' with those two columns and finding the correlation between them

Percentile_rank = GSVA_states['Per Capita GSDP (Rs.)'].rank(pct=True)
Primary_rank = GSVA_states['Primary_Percent'].rank(pct=True)
GSVA_states['Percentile_rank'] = GSVA_states['Per Capita GSDP (Rs.)'].rank(pct=True)
GSVA_states['Primary_rank'] = GSVA_states['Primary_Percent'].rank(pct=True)
corr_df = pd.DataFrame({'Percentile_rank': GSVA_states.iloc[:,-2],'Primary_rank': GSVA_states.iloc[:,-1]})
corr_df.corr()
```

__Categorise the states into four groups based on the GDP per capita (C1, C2, C3, C4, where C1 would have the highest per capita GDP and C4, the lowest)__

```python
# The quantile values are (0.20,0.5, 0.85, 1)
# The states lying between the 85th and the 100th percentile are in C1
# Those between the 50th and the 85th percentiles are in C2, and so on
# Creating a new column 'Category' in the original data frame

GSVA_states['Category'] = pd.qcut(GSVA_states['Percentile_rank'],[0,0.2,0.5,0.85,1],['C4','C3','C2','C1'],precision = 100)
GSVA_states
```
